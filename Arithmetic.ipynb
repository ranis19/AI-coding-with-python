```python
import mymath

num1=float(input("Enter first number: "))
num2=float(input("Enter second number: "))
operator=input("Enter the operator: add,subtract,multiply,division? ").strip().lower()

if operator == "add":
    print("Result is", mymath.add(num1,num2))
elif operator == "subtract":
    print("Result is", mymath.subtract(num1,num2))
elif operator == "multiply":
    print("Result is", mymath.multiply(num1,num2))
elif operator == "division":
    print("Result is", mymath.division(num1,num2))
else:
    print("It is invalid.")
    
```
