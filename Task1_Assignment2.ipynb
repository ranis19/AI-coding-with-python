```python
supplies = ["Pens", "Pencils", "Notebooks"]

quantities = {"Pens": 10, "Pencils": 20, "Notebooks": 5}

categories = {"Writing", "Stationery"}

a=int(input("How many more pens do you want?"))
quantities["Pens"]+=a
print(quantities)  #Task1.1
```

```python
q_pencils=quantities["Pencils"]
q_notebooks=quantities["Notebooks"]
if q_pencils > q_notebooks:
    print("the quantity of Pencils is greater than the quantity of Notebooks.")
elif q_pencils == q_notebooks:
    print("the quantity of Pencils is same as the quantity of Notebooks.")
else:
    print("the quantity of Pencils is less than the quantity of Notebooks.")
 #Task1.2
```

```python
if q_pencils>10 and q_notebooks<10:
    print ("The quantity of Pencils is greater than 10 while the quantity of Notebooks is less than 10.")
else:
    print ("We don't know whether the quantity of Pencils is greater than 10 or the quantity of Notebooks is less than 10.") 

#Task1.3
```

```python
def add_supplies():
    x = list(quantities.values())
    y = list(quantities.keys())
    for n in range(len(x)):
        x[n]=x[n]*1.25
        quantities[y[n]]=x[n]
    print(quantities)

add_supplies()  #Task3.1    
```

```python
def show_totals():
    add_supplies()
    x = list(quantities.values())
    total = 0
    for n in range(len(x)):
        total += x[n]
    print("The total quantities are",total)
show_totals()  #Task3.2
```
