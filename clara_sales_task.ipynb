```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#2. Generate and organize data
sales_data = np.array([
    [120, 250, 100],
    [130, 160, 110],
    [125, 170, 105]
])
months = ['Jan', 'Feb', 'Mar']
products = ['product A', 'product B', 'product C']
df=pd.DataFrame(data=sales_data,index=months, columns=products)
print(df)

#3. Display Basic Statistics using df.describe()
summary = df.describe()
print(summary)
#mean is the average of the values
#min is the smallest value
#count is the number of data

#4. Visualization and add markdown explanation of the chart
plt.plot(df.index, df['product A'], marker='o', label='Product A')
plt.plot(df.index, df['product B'], marker='x',label='Product B')
plt.plot(df.index, df['product C'], marker='^',label='Product C')
plt.title('Monthly Sales Performance')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend() 
#product B shows consistant growth.

#% Download the notebook as clara_sales_task.ipynb
```
