```python
import area_utils

shape=input("Choose a shape:circle,rectangle,triangle  \n").strip().lower()
if shape == "circle":
    radius=float(input("Enter the radius: "))
    print("The area of circle is",area_utils.area_circle(radius))
elif shape == "rectangle":
    w = float(input("Enter the width "))
    l = float(input("Enter the length "))
    print("The area of rectangle is",area_utils.area_rectangle(w,l))
elif shape == "triangle":
    b = float(input("Enter the bass"))
    h = float(input("Enter the height"))
    print("The area of triangle is",area_utils.area_triangle(b,h))
else:
    print("It's valid.")
    
```
