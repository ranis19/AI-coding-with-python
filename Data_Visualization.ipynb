```python
#Question 1: Histogram of Exam Scores
#Plot a histogram of the exam_score column to visualize the distribution of scores. 
#Hint: Use plt.hist() and label the axes.

import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv(r"C:\Users\Clara\OneDrive - South Thames Colleges Group\Desktop\a18dec_work@h\students_performance.csv")
plt.title("Histogram of Exam Scores")
plt.xlabel("Exam_Score")
plt.ylabel("Frequency")
plt.hist(df["exam_score"],color='green',alpha=0.3,bins=50)
plt.show() # %matplotlib inline 
```

```python
#Question 2: Bar Chart of Average Scores by Class Level
#Create a bar chart showing the average exam score for each class_level(Year 10, Year 11, Year 12). 
#Hint: Use groupby() and plot(kind="bar").

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\Clara\OneDrive - South Thames Colleges Group\Desktop\a18dec_work@h\students_performance.csv")
result=df.groupby("class_level")["exam_score"].mean()
min_val = result.min()
max_val = result.max()
plt.ylim(min_val - 5, max_val + 5) # can show more celar gap results
result.plot(kind="bar", color='teal', alpha=0.3, edgecolor='black',width=0.4) #the data plot with 0.4 width
plt.xlabel=('class_level')
plt.ylavel=('exam_score')
plt.xticks(rotation=0) # rotate numbers in X axis in 90 degreee clockwise
plt.title('Average Exam Score by Class Level')
plt.show()
```

```python
#Question 3: Scatter Plot of Hours Studied vs Exam Score
#Make a scatter plot of hours_studied vs exam_score to explore the relationship 
#between study time and performance. 
#Hint: Use plt.scatter() and add axis labels.
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(9, 6)) #(xaxis,yaxis)
df=pd.read_csv(r"C:\Users\Clara\OneDrive - South Thames Colleges Group\Desktop\a18dec_work@h\students_performance.csv")
plt.scatter(df["hours_studied"], df["exam_score"], color='teal', alpha=0.8, s=5) #s is the radius of dots
plt.title("Study time and performance")
#plt.xlabel("studied hour")
#plt.ylabel("exam_scores")
plt.show()
```

```python
#Question 4: Boxplot of Exam Scores by Gender
#Draw a boxplot comparing exam_score distributions for Male and Female students. 
#Hint: Use sns.boxplot() with x="gender" and y="exam_score".
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\Clara\OneDrive - South Thames Colleges Group\Desktop\a18dec_work@h\students_performance.csv")
plt.figure(figsize = (5, 5))
sns.boxplot(x='gender',y='exam_score',data=df, width=0.5,whis=3) #whis is the length of whiskers
plt.ylim(0, 105) #extend y axis 
plt.xlabel("gender")
plt.ylabel("exam_scores")
plt.show()
```

```python
#Question 5: Line Plot of Average Exam Score by Exam Month
#Create a line plot showing how the average exam_score changes across exam_month. 
#Hint: Sort months in order and use plt.plot().
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\Clara\OneDrive - South Thames Colleges Group\Desktop\a18dec_work@h\students_performance.csv")
months_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df['exam_month'] = pd.Categorical(df['exam_month'], categories=months_order, ordered=True)
df_sorted = df.groupby('exam_month')['exam_score'].mean()
plt.figure(figsize=(10,5))
df_sorted.plot(kind='line',color="green",alpha=0.4,linewidth=2)
plt.xlabel("exam_month")
plt.ylabel("exam_score")
plt.title("Line Plot of Average Exam Score by Exam Month")
plt.show()
```
